```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
db = DBSCAN(eps=3,min_samples=2)  # eps means radius and min_samples means circumference of circle
db.fit(X)
db.labels_

# 0 ,0,0 means one cluster and 1,1 means another clusters and -1 represent noise points or outliers
```
