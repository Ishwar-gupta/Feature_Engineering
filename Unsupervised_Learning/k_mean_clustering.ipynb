```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing the dataset
import pandas as pd

df = pd.read_csv("student_clustering.csv", encoding='latin-1')
df.head(5)
```

```python
# the shape of the dataset
df.shape
```

```python
# plotting scatter plot of the dataset
plt.scatter(df['cgpa'],df["iq"])
```

```python
# finding k using elbow method

from sklearn.cluster import KMeans

# making empty list of wcss

wcss =[]

# using for loop for repeatedly finding k
for i in range(1,11):
    #making object of kmeans
    km =KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
#plotting the elbow graph to exaclty know the value of k
plt.plot(range(1,11),wcss)
```

```python
X = df.iloc[:,:].values
km=KMeans(n_clusters=4)
y_mean = km.fit_predict(X)
```

```python
"""
km.fit_predict(X)

This is the most important line.

It performs two tasks together.

1. fit(X)

Train the K-Means model.

During training, K-Means

Chooses 4 random centroids.
Calculates the distance of every point from each centroid.
Assigns every point to the nearest centroid.
Recalculates the centroid.
Repeats until the centroids stop changing


2. predict(X)

After training, predict which cluster each data point belongs to.
"""
```

### K-Means on 3-D data


```python
from sklearn.datasets import make_blobs

centriods = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200 , cluster_std=cluster_std,centers= centriods , n_features=3,random_state=1)
```

```python
X
```

```python
import plotly.express as px

fig = px.scatter_3d(x=X[:,0],y=X[:,1],z=X[:,2])
fig.show()
```

```python
wcss =[]

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_) # .inertia_ means is used when implementing the Elbow Method to find the optimal number of clusters (k) for K-Means.
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=4)

y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]

df['label'] = y_pred
```

```python
fig = px.scatter_3d(df,x='col1',y='col2',z='col3',color='label')
fig.show()
```
