```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep= 20)

# n_features => Number of useful features that actually help predict the target.
# n_redundant => No copied/correlated features are added.
# n_clusters_per_class => Number of groups (clusters) inside each class. Each class forms only one cluster.
# random_state =41 means it control the randomness for reproducibilty and Every time you run code, same dataset is generated.
# class_sep =10 -> Large value ⇒ classes far apart ⇒ easier classification.
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c = y ,cmap="winter",s=100)
plt.show()
```

### perceptron function using step()

```python
def perceptron(X,y):

    X=np.insert(X,0,1,axis=1)  #adding bias term  insert 1 at 0th index on X feature
    weights = np.ones(X.shape[1]) # makes ones of two column
    lr=0.1 #learning_rate=0.1

    for i in range(1000):
        j = np.random.randint(0,100) # Chooses a random training point index 
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr * (y[j] - y_hat)*X[j]
    return weights[0],weights[1:] 

def step(z):
    return 1 if z>0 else 0
# '''
# z > 0  → class 1
# z <= 0 → class 0
# '''
    
```

```python
intercept_,coef_=perceptron(X,y)

print("Intercept_:",intercept_)
print("coef_:",coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[0])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m * x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth =3)
plt.scatter(X[:,0],X[:,1],c=y , cmap='winter',s=100)
plt.ylim(-5,5)
plt.show()
```
