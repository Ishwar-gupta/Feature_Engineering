```python
# Importing required libraries  

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### loading datasets for classification visualization

```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=30)
```

### visualization using scatterplot


```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=50)  
'''
# c=y meansColor points according to target labels.

Graph may show:
Blue points → class 0
Green points → class 1

cmap means colormap.

It decides WHICH colors to use.

s means size of points.  s=20 means small dots and s=200 means larger dots
'''

plt.figure(figsize=(10,6))
plt.show()
```

```python
def perceptron(X,y):

    X=np.insert(X,0,1,axis=1)  #adding bias term  insert 1 at 0th index on X feature
    weights = np.ones(X.shape[1]) # makes ones of two column
    lr=0.1 #learning_rate=0.1

    for i in range(1000):
        j = np.random.randint(0,100) # Chooses a random training point index 
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr * (y[j] - y_hat)*X[j]
    return weights[0],weights[1:] 

def step(z):
    return 1 if z>0 else 0
# '''
# z > 0  → class 1
# z <= 0 → class 0
# '''
    
```

```python
intercept_,coef_=perceptron(X,y)

print("Intercept_:",intercept_)
print("coef_:",coef_)
```

```python
m = -(coef_[0]/coef_[1])  # m = - x_coef/y_coef
b = -(intercept_/coef_[1]) # c = - intercept/y_coef
```

```python
x_input = np.linspace(-3,3,100)
y_input =  m * x_input + b 
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input ,color='red',linewidth = 3)
plt.scatter(X[:,0],X[:,1],c=y,cmap="winter",s=100)
plt.ylim(-5,5)

# """
# control visible y-axis range
# y-axis starts at -3
# y-axis ends at 2
# """
```

### Using direclty Logistics Regression

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(X,y)
```

```python
# calculation of m and b
m = -(log.coef_[0][0] / log.coef_[0][1])
b = -(log.intercept_ /log.coef_[0][1])
```

```python
# calculation of x_input and y_input for logistic reg

x_input1 = np.linspace(-3,3,100)
y_input1 = m * x_input1 + b
```

## visualization of perceptron vs logistics regression

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth = 3)
plt.plot(x_input1,y_input1 , color ='black',linewidth = 1)
plt.scatter(X[:,0],X[:,1], c = y , cmap='winter',s=100)
plt.ylim(-5,5)
plt.show()
```
